```python
import pandas as pd

#load the data raw and normalized
Data_raw_and_normalized = pd.read_csv("../Data/TephraDataSet_raw_and_normalized.csv", encoding = 'latin1', low_memory =False )
```

 In the dataset the data was stored as in the original publication. Because of this, in many cases the assigned value to a particular element might not be a number but an expresion to indicate: missingness, lower than a certain value, below detection limit (bdl) or over range. In order to treat the data these values are replaced by NAN. In order to process the data it is necesary to replace these expressions by "NaN".

```python
import numpy as np
Data_raw_and_normalized = Data_raw_and_normalized.replace(to_replace='Over range',value=np.nan)
Data_raw_and_normalized = Data_raw_and_normalized.replace(to_replace='bdl',value=np.nan)
Data_raw_and_normalized = Data_raw_and_normalized.replace(to_replace='<0.01', value=np.nan)
Data_raw_and_normalized = Data_raw_and_normalized.replace(to_replace='<0.1',value=np.nan)
Data_raw_and_normalized = Data_raw_and_normalized.replace(to_replace='<1',value=np.nan)
Data_raw_and_normalized = Data_raw_and_normalized.replace(to_replace='<5',value=np.nan)
Data_raw_and_normalized = Data_raw_and_normalized.replace(to_replace='<6',value=np.nan)
Data_raw_and_normalized = Data_raw_and_normalized.replace(to_replace='<10',value=np.nan)
Data_raw_and_normalized = Data_raw_and_normalized.replace(to_replace='n.a.',value=np.nan)
Data_raw_and_normalized = Data_raw_and_normalized.replace(to_replace='Not analyzed',value=np.nan)
Data_raw_and_normalized = Data_raw_and_normalized.replace(to_replace='-',value=np.nan)
Data_raw_and_normalized = Data_raw_and_normalized.replace(to_replace='Not Netermined',value=np.nan)
```

Check how much the normalized values of each element changes when renormalizing the data by a new 
Total = SiO2 + TiO2 + Al2O3 + MgO + CaO + Na2O + K2O (i.e. without MnO, P2O5 or CL)

Only data where all major elements have been analyzed (corresponding to 2995 Sample Observations from Case 1) are tested.

```python
case1_renormalizado = case1.copy()
case1_renormalizado = case1_renormalizado.reset_index()
case1 = case1.reset_index()
case1_renormalizado['MnO'] = case1_renormalizado['MnO'].replace('-',-1).astype(float)
case1_renormalizado['P2O5'] = case1_renormalizado['P2O5'].replace('-',-1).astype(float)
case1_renormalizado['Cl'] = case1_renormalizado['Cl'].replace('-',-1).astype(float)

#Defining some variables which we will plot later to understand the variability of the re normalized data 
case1_renormalizado['MnO + P2O5 + Cl'] = 'default'
case1_renormalizado['Analytical Total without LOI'] = 'default'

for i in range(0,len(case1_renormalizado.Total)):
    sum_ = np.nansum([case1_renormalizado.SiO2[i],
                      case1_renormalizado.TiO2[i],
                      case1_renormalizado.Al2O3[i],
                      case1_renormalizado['FeOT'][i],
                      case1_renormalizado.MgO[i],
                      case1_renormalizado.CaO[i],
                      case1_renormalizado.Na2O[i],
                      case1_renormalizado.K2O[i]])
    
    case1_renormalizado.loc[i,'MnO + P2O5 + Cl'] = np.nansum([case1_renormalizado.MnO[i],
                                                              case1_renormalizado.Cl[i],
                                                              case1_renormalizado.P2O5[i]])
    case1_renormalizado.loc[i,'Analytical Total without LOI'] = np.nansum([case1_renormalizado.Total[i],
                                                                           -case1_renormalizado.LOI[i]])
    
    for elemento in ['SiO2','TiO2','Al2O3','FeOT','MgO','CaO','Na2O','K2O']:
        case1_renormalizado.loc[i,elemento] = case1_renormalizado[elemento][i]*100/sum_
```

```python
#load MeasurementRuns file to now the secondary standards measured along with the samples
Data_standards = pd.read_excel('../Data/MeasurementRuns.xlsx')
Data_standards.loc[:,'MeasurementRun'] = Data_standards.loc[:,'MeasurementRun'].astype('str')
case1_renormalizado.loc[:,'MeasurementRun'] = case1_renormalizado.loc[:,'MeasurementRun'].astype('str')

#load the Secondary standards reference values
Standards_ref = pd.read_excel('../Data/Standards_Reference.xlsx')
Standards_ref = Standards_ref[Standards_ref.error.isin(['95%CL','SD'])].copy()
```

```python
#get which secondary standards where analyzed along with the samples
stds =  Data_standards[Data_standards.MeasurementRun.isin(case1_renormalizado.MeasurementRun.unique())].StandardID.unique()
Standards_ref_case1 = Standards_ref[Standards_ref.StandardID.isin(stds)]
print(Standards_ref_case1.StandardID.unique())
```

```python
#obtaining the difference between the original normalized values (with Totals taking into consideration different elements)
#and the re-normalized values
case1_renormalizado['SiO2'] = np.subtract(case1_renormalizado['SiO2'],case1['SiO2_normalized'])
case1_renormalizado['TiO2'] = np.subtract(case1_renormalizado['TiO2'],case1['TiO2_normalized'])
case1_renormalizado['Al2O3'] = np.subtract(case1_renormalizado['Al2O3'],case1['Al2O3_normalized'])
case1_renormalizado['FeOT'] = np.subtract(case1_renormalizado['FeOT'],case1['FeOT_normalized'])
case1_renormalizado['MgO'] = np.subtract(case1_renormalizado['MgO'],case1['MgO_normalized'])
case1_renormalizado['CaO'] = np.subtract(case1_renormalizado['CaO'],case1['CaO_normalized'])
case1_renormalizado['Na2O'] = np.subtract(case1_renormalizado['Na2O'],case1['Na2O_normalized'])
case1_renormalizado['K2O'] = np.subtract(case1_renormalizado['K2O'],case1['K2O_normalized'])
```

```python
import plotly.express as px
import plotly.graph_objects as go

elementos = ['SiO2','TiO2','Al2O3','FeOT','MgO','CaO','Na2O','K2O']

for elemento in elementos: #['SiO2','Al2O3']:
    plot1 = go.Scatter(x=case1_renormalizado[elemento+'_normalized'],
               y=case1_renormalizado[''+elemento],
               mode = 'markers',
               hovertext = 'Sample Observation ID: ' + case1_renormalizado.SampleObservationID.astype('str') + ' '+ 
                       case1_renormalizado.Authors.astype('str') ,
               marker = dict(color=case1_renormalizado['MnO + P2O5 + Cl'],showscale = True),
               name = "MnO + P2O5 + Cl")

    plot2 = go.Scatter(x = Standards_ref_case1[elemento],
                  y = Standards_ref_case1['_'+elemento],
                  text = Standards_ref_case1.StandardID ,
                  hovertext = 'Secondary standard: ' + Standards_ref_case1.StandardID  , 
                  mode = 'markers+text',
                  textposition = 'top center' ,
                  showlegend = False ,
                  line = {'color': "#3b3b3b"})

    Data = [plot1,plot2]
    layout = go.Layout(xaxis=dict(title=elemento), yaxis=dict(title=''+elemento),
                      legend={'y':1.1, 'x':0.98}, width=800, height=500)# title='Re-normalization variation',
    
    fig = go.Figure(data=Data,layout=layout)
    #fig.show("png", engine='kaleido')

    fig.write_image('../Plots/temp.png')
```
